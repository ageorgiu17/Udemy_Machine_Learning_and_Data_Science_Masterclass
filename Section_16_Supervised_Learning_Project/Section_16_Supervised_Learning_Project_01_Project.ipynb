```python
def cohort(tenure):
    if tenure < 13:
        return '0-12 Months'
    elif tenure < 25:
        return '12-24 Months'
    elif tenure < 49:
        return '24-48 Months'
    else:
        return 'Over 48 Months'
```

```python
df['Tenure Cohort'] = df['tenure'].apply(cohort)
```

```python
df[['Tenure Cohort', 'tenure']]
```

```python
X = df.drop(['Churn', 'customerID'], axis=1)
y = df['Churn']
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(max_depth=6)
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import plot_confusion_matrix, classification_report
```

```python
y_pred = model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
plot_confusion_matrix(model, X_test, y_test)
```

```python
imp_features = pd.DataFrame(index=X.columns, data=model.feature_importances_, columns=['Importances'])
imp_features = imp_features.sort_values('Importances')
imp_features = imp_features[imp_features['Importances'] > 0]
```

```python
sns.barplot(data=imp_features, y='Importances', x=imp_features.index)
plt.xticks(rotation=90);
```
