# Matplotlib - Overview and Basics

Vizualizarea datelor este crucială pentru a înțelege relațiile din data setul cu care se lucrează. Librăria 'Matplotlib' este una dintre cele mai populare librării care se ocupă cu partea de vizualizare de date. Majoritatea librăriilor de vizualizare (Seaborn) sunt construite având în spate librăria Matplotlib. Un aspect pozitiv al aceste librării este faptul că se poate construi aproape orice plot cu o parte de customatizare extrem de bună. Această parte de customatizare face un pic mai greu partea de învățare a sintaxei de cod. Un alt motiv pentru sintaxa mai complexă este faptul că există două metode separate de a crea plot-uri, pe baza funcțiilor și pe baza claselor.

Cel mai simplu mod de a utiliza Maptplotlib este prin apelarea funcțiilor de plot (plt.plot()). Pentru a putea utiliza această metodă o să fim nevoiți să importăm librăria de matplotlib. Din librăria respectivă trebuie să importăm modulul de pyplot (matplotlib.pyplot). Prin convenție, acesta se importă sub denumirea de 'plt'

```python
import matplotlib.pyplot as plt
```

Pentru început o să creem anumite ploturi pentru array-uri, cele mai simple ploturi. Pentru asta o să ne folosim de librăria numpy pentru a crea array-urile respective

```python
import numpy as np
```

```python
x = np.arange(0, 10)
```

```python
y = x * 2
```

Cel mai simplu mod de a vizualiza aceste date este să utilizăm metoda 'plt.plot()'. Acestei metode o să îi oferim ca și argumente array-urile x și y

```python
plt.plot(x, y)
```

Prin rularea acelei comenzi, matplotlib a creat un plot liniar pentru acele valori. Ce anume putem modifica la acest plot? Pentru început putem să aîi adăugăm un titlu acestui plot. Pentru a face asta o să utilizăm metoda plt.title(). Ca și argument o să îi atribuim un string care o să reprezinte titlul acestui plot

```python
plt.plot(x, y)
plt.title('Linear plot between x and y')
```

Următoarea modificare care se poate face este de a crea label-uri pentru axa x și y. Matplotlib ne pune la dispoziție două metode, și anume plot.xlabel() și plot.ylabel(). Ca și argument au nevoie de un string care o să reprezinte label-ul pentru axa respectivă

```python
plt.plot(x, y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Linear plot between X and Y')
```

În plot-urile de mai sus se poate observa faptul că linia respectivă nu începe chiar de la baza graficului, ci are o zonă în care nu este linie (unde valoarea este teoretic sub 0). Ce anume se poate face pentru a renunța la această zonă, este să ne setăm limitele pentru axele graficului. Pentru a face asta o să utilizăm metodele plot.xlim(), plot.ylim(). Ca și argument au nevoie de două valori, prima care reprezintă valoarea de început pentru limită, iar a doua, valoarea de final. O să modificăm limta de la axa X pentru a merge de la 0 până la 6

```python
plt.plot(x, y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xlim(0, 6)
plt.title('Linear plot between X and Y')
```

În cazul în care se dorește, aceste plot-uri care sunt create de matplotlib pot fi salvate ca și fișiere (png, jpeg) pe mașina de lucru. Pentru a le salva o să utilizăm metoda 'plt.savefig()'. În ceea ce privește argumentul de care are nevoie, trebuie să îi oferim un string în care să fie trecut numele pe care dorim să îl oferim acestui plot ce o să fie salvat. Dacă se oferă doar numele, atunci plot-ul o să fie salvat în directorul unde se rulează notebook-ul. Dacă se dorește, înainte de nume se poate specifica și locația unde anume să se salveze acest plot

plt.savefig('plot.png')
