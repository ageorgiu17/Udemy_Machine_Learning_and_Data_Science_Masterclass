# Matplotlib - Styling Legends

Până în acest moment am creat diferite plot-uri în cadrul acestui tutorial. În continuare o să mai creem încă un obiect, dar pentru un set de axe o să rulăm de două ori metoda plot pentru a vedea ce anume se întâmplă când facem asta

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 10)
```

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, x)
axes.plot(x, x ** 2)
```

După cum se poate observa, dacă am utilizat de două ori metoda plot pentru un set de axe s-au creat două linii care arată ce relație există între datele respective. Ce nu putem să ne dăm seama din aceste linii însă este ce anume reprezintă liniile respective. Pentru a putea înțelege ce anume reprezintă, o să trebuiască să creem o legengă pentru acest grafic. Pentru asta trebuie să facem doi pași. Primul este ca în cadrul metodei 'plot' să utilizăm un argument denumit label, argument care o să reprezinte defapt ce anume o să se afișeze pe legenda din cadrul graficului la final.

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, x, label=' X vs X')
axes.plot(x, x ** 2, label='X vs X ^ 2')
```

Următorul pas pentru a ne afișa legenda în cadrul acestui grafic este să apelăm metoda '.legend()' pe setul de axe pe care am creat acest plot (Aceste plot-uri în cazul de față)

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, x, label=' X vs X')
axes.plot(x, x ** 2, label='X vs X ^ 2')

axes.legend()
```

Prin utilizarea metodei respective putem observa că acuma legenda apare pe grafic, iar fiecare linie este afișată cu valoarea care a fost trecută pentru argumentul de label

## Recapitulare

În cadrul acestui tutorial am învățat:

    1. Cum să adăugăm un label pentru un plot

        fig = plt.figure()

        axes = fig.add_axes([0, 0, 1, 1])

        axes.plot(x, y, label='plot for x and y variables')

    2. Cum să adăugăm legenda pentru un canvas

        axes.legend()
