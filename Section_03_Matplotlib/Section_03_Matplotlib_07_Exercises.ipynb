# Matplotlib - Exercises

1. Crează două array-uri numpy prin care să expunem relația E = mc^2

```python
import numpy as np
```

```python
m = np.linspace(0, 10, 11)
```

```python
m
```

```python
E = m * (300_000 ** 2)
```

```python
E
```

2. Creați un plot pentru ecuația respectivă

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(m, E)
plt.xlabel('Mass in grams')
plt.ylabel('Energy in Joules')
plt.title('E = mc^2')
```

3. Pentru ecuația de mai sus creați un plot logaritmic

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.set_yscale('log')
axes.plot(m, E)
```

În continuare se dau următoarele date:

```python
labels = ['1 Mo', '3 Mo', '6 Mo', '1 Y', '2 Y', '3 Y', '5 Y', '7 Y', '10 Y', '20 Y', '50 Y']

july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

4. Să se reprezinte datele respective pe aceeași figură. Adăugați și o legendă pentru aceste ploturi


```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(labels, july16_2007, label='july16_2007')
axes.plot(labels, july16_2020, label='july16_2020')
axes.legend()
```

5. Afișați cele două grafice pe două ploturi 

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(15, 7), dpi=300)
axes[0].plot(labels, july16_2007)
axes[1].plot(labels, july16_2020)
```

6. Bonus! Creati un plot cu axe twin

```python
fig, axes1 = plt.subplots()
axes1.plot(labels, july16_2007, color = 'blue')

axes2 = axes1.twinx()
axes2.plot(labels, july16_2020)
```
