# Matplotlib - Styling Colors and Styles

După ce am învățat cum anume să adăugăm o legendă pentru un anumit plot, urmează să învățăm cum anume putem modifica culorile și stilul unui plot. Aici intră modificări precum stilul liniei, culoare liniei respective, mărimea liniei. Să începem să implementăm aceste noțiuni în cadrul unui plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 11)
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, x)
```

Până în acest moment am învățat cum să modificăm un set de axe, cum să îi adăugăn un anumit titul, label-uri pentru axe, limite pentru axele respective. În acest moment nu știm cum să modificăm acea linie care este generată ca și plot. O să începem prin a modifica culoarea linii respective. Pentru a modifica culoarea respectivă o să utilizăm pentru metoda plot() argumnetul 'color'. Ceea ce putem să îi oferim acestui parametru este fie un string care reprezintă numele culorii pe care dorim să o oferim (blue, lightblue, purple) fie un cod Hex RGB (acesta se poate căuta pe internet)

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, x, color='#145926')
```

Pentru cazurile în care dorim să modificăm mărimea acelei linii, pentru asta o să ne folosim de parametrul 'linewidth'. Acestui parametru o să îi oferim ca și valoare un integer care reprezintă defapt o rație în comparație cu dimensiunea standard. Valorea este 1 setată default, iar dacă îi setăm o nouă valoare de 3, atunci nou linie o să fie mai mare de 3 ori precum linia care este creată default

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, x, color='#145926', linewidth=3)
```

Stilul liniei se modifică prin parametrul 'linestyle'. Acestuia o să îi oferim un string ca și valoare în care se specifică se stil nou dorim să aibă linia respectivă. Variantele respective pot fi:

    '--'

    '-.'

    '-'

    ':'

Pentru mai multe stiluri se poate face un serach pe google

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, x, color='#145926', linewidth=3, linestyle='-.')
```

Ideea de bază este că linia care se afișează reprezintă corelația dintre datele pe care le avem. Datele respective sunt defapt coordonate (puncte), dar acestea nu ne sunt afișate în cadrul acestui grafic. Pentru a afișa punctele resective există parametrul 'marker' ce are nevoie de o valoare de tip string prin care se specfifică ce tip de marker să fie afișate pentru punctele respective. Lista cu markerele disponibile se poate găsi la link-ul https://matplotlib.org/stable/api/markers_api.html

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, x, color='#145926', linewidth=3, linestyle='-.', marker='o')
```

Mărimea markerelor se va seta utilizând parametrul de 'markersize' (o să îi atribuim un integere ca și valoare care reprezintă mărimea în puncte (points per dpi))

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, x, color='#145926', linewidth=3, linestyle='-.', marker='o', markersize=10)
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, x, color='#145926', linewidth=3, linestyle='-.', marker='o', markersize=10)
axes.plot(x, x + 1, color='red', linewidth=2, marker='8', markersize=4)
axes.plot(x, x - 1, color='lightblue', linewidth=0.8, label='label for thrid plot')

axes.legend()
```
