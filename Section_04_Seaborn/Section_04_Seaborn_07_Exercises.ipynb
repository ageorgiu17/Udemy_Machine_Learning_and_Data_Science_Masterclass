# Seaborn Exercises

```python
# Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the csv file into a DataFrame
df = pd.read_csv('../data/05-Seaborn/application_record.csv')
```

```python
# printing the head of the DataFrame
df.head()
```

```python
df.shape
```

```python
employed = df[df['DAYS_EMPLOYED'] < 0]
```

```python
employed.shape
```

```python
employed['DAYS_BIRTH'] = employed['DAYS_BIRTH'].abs()
employed['DAYS_EMPLOYED'] = employed['DAYS_EMPLOYED'].abs()
```

```python
plt.figure(figsize=(10, 4))
sns.scatterplot(x='DAYS_BIRTH', y='DAYS_EMPLOYED', data=employed, alpha=0.02, linewidth=0)
```

```python
employed['YEAR'] = employed['DAYS_BIRTH'] / 365
```

```python
sns.displot(x='YEAR', data=employed, bins=40)
```

```python
df['YEAR'] = df['DAYS_BIRTH'].abs() / 365
```

```python
sns.displot(x='YEAR', data=df, bins=40)
```
