# Seaborn - Scatterplots

Un scatterplot arată realația dintre două feater-uri de tip 'continuous'. Un element de tip continuous este un element care poate lua orice număr de valori între două valori stabilite. Aceste elemte pot fi valori care se tot pot împărți, cum ar fi vârsta unei persoane, înălțimea, salariul, temperatura, prețul unui obiect. De exemplu, pentru vârsta unei persoane se poate spune că are 24 de ani. Dacă dorim să fim mai preciși putem spune că are 24 de ani și 8 luni. Tot așa se pot adăuga date până ce se ajunge la vârsta unei persoane exprimate chiar și în secunde.

Pe lână aceste valori, mai există și valorile categorice, precum culoare, nume, formă, care sunt valori prestabilite ce nu mai pot fi divizate.

Un scatterplot ia două valori continue și le afișează sub formă de coordonate într-un plot. O să începem să folosim aceste idei în cadrul codului. Ca și convenție, s-a stabilit ca librăria Seaborn să se importe sub denumirea de sns

```python
# importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the data into a DataFrame
df = pd.read_csv('../data/05-Seaborn/dm_office_sales.csv')
```

```python
# Listing the first five elements from the DataFrame
df.head()
```

În cadrul DataFrame-ului de mai sus avem mai multe date referitoare la anumite persoane ce lucrează la partea de vânzări, date precum divizia din care fac parte, nivelul educației, experiența de muncă în ani, salariul persoanei și numărul de vânzări realizate de către acea persoană. Poate ni se spune să aflăm ce realație există între numărul de vânzări și salariul unei persoane din companie. Cel mai ușor mod de a vizualia ce tip de relație există între aceste este vizual, iar aici intră în discuție librăria Seaborn.

De cele mai multe ori o să apleăm librăria respectivă (sns) asupra căreia o să utilizăm un anumit plot. Dacă se aplelează libraia respectivă și se notează un punct după, o să ni se afișeze ca și un help în Jupyter Notebook toate ploturile disponibile (există și anumite atribute sau anumite funcții care nu sunt ploturi) din această librărie. Pentru început o să utilizăm un scatterplot

```python
sns.scatterplot()
```

Acestui plot o să îi atribuim ceea ce are nevoie pentru axa x și y și de unde provin datele respective (din ce DataFrame). Pentru valori pentru parametrii x și y o să se ofere câte un string care reprezintă numele coloanei din DataFrame-ul unde avem datele.

```python
sns.scatterplot(x='salary', y='sales', data=df)
```

Seaborn creează acel plot utilizând în spate Matplotlib. Din acest motiv putem să extragem obeictul de tip feagure și să îl edităm precum am învățat în secșiunea de Matplotlib. În aceeași celulă în care am aplelta plot-ul scatterplot, înainte putem să utilizăm comenzi de Matplotlib pentru a modifica acel obiect

```python
plt.figure(figsize=(8, 4), dpi=300)
sns.scatterplot(x='salary', y='sales', data=df)
```

Ce anume putem să mai modificăm la acest scatterplot? Acest tip de plot are un parametru denumit 'hue' care acceptă ca și argument o coloană. Datele din acea coloană reprezintă cum să fie colorate aceste puncte din cadrul plotului. Momenta acestea sunt colorate toate în aceeași culoare (albastru). Dacă dorim să afișăm datele în funcție de nivelul de educație al fiecărei persoane, atunci putem să îi atribuim acea coloană parametrului 'hue', iar astfl o să coloreze fiecare punct într-o culoare specifică în funcție din ce categorie face parte. Acest parametru funcționează cel mai bine cu o coloamă de tipul Categorical unde avem mai multe categorii

```python
plt.figure(figsize=(8, 4), dpi=300)
sns.scatterplot(x='salary', y='sales', data=df, hue='level of education')
```

Ceea ce este foarte util la acest parametru este faptul că ne și afișează legenda în cadrul plot-ului respectiv. Spuneam că acest parametru funcționează cel mai bine cu o coloană de tip categorical, dar acest parametru poate funcționa și cu o coloană de tip continuous deoarece Seaborn în spate generează intervale pentru aceste date și creează un gradient prin care să se diferențieze aceste grupe de date

```python
plt.figure(figsize=(8, 4), dpi=300)
sns.scatterplot(x='salary', y='sales', data=df, hue='salary')
```
