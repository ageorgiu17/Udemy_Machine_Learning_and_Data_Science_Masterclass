# Seaborn - Distribution plots

Aceste tipuri de ploturi afișează datele pentru un singur feature dintr-o coloană de tip continuous și ne ajută să aflăm informații precum deviația sau valoarea medie. Există mai multe astfel de tipuri precum:

    - Rug Plot

    - Histogram

    - KDE Plot

Un Rog plot este cel mai simplu plot de distribuție, iar acesta doar adaugă câte o linie pentru valorile dintr-o coloană. Axa y nu are niciun rost în cadrul acestui plot. Problema cu acest tip de plot este faptul că axa y nu se poate interpreta, ce se poate observa din cadrul acestor plot-uri ar fi prezența unor outliers. Să vizualizăm aceste date pentru un fișier csv

```python
# importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reading the csv file into a DataFrame
df = pd.read_csv('../data/05-Seaborn/dm_office_sales.csv')
```

```python
# printing the head of the DataFrame
df.head()
```

Pentru a crea un plot de tip Rug plot o să utilizăm comanda 'sns.rugplot()'. De obicei pentru un plot trebuie să specificăm date pentru axele x și y, dar dupăcum am spus, pentru un plot de tipul rug plot, axa y nu are niciun înțeles, din acest motiv este destul să oferim date pentru parametrii 'x' și 'data'

```python
sns.rugplot(x='salary', data=df)
```

După cum am spus, acest plot pune câte o linie pentru ficare valoare din cadrul coloanei de 'salary'. Pentru a face aceste linii mai mari putem să ne folosim de parametrul 'height' pentru care să îi oferim un procent. (1 înseamnă să ocupe toate înălțimea de la axa y, iar 0 înseamnă să se afișeze nimic)

```python
plt.figure(dpi=200)
sns.rugplot(x='salary', data=df, height=0.32)
```

Din acest plot ce se poate observa este faptul că există anumiți outliers, adică există date care sunt separate de restul la începutul și sfârșitul graficului, date care ar putea să modifice anumite calcule sau anumite statistici

Din moment ce acest tip de plot doar afișează o linie pentru fiecare valoare dintr-o anumită coloană, este destul de greu să știm dacă există sau nu valiori duplicate. Pentru a afla distribuția valorilor mai precis se va utiliza o Histogramă. Ce anume face histograma, împarte valorile dintr-o anumită coloană în mai multe grupe și numără câte valori fac parte dintr-o anumită grupă. Numărul valorilor care apar per grupă o să fie afișate pe axa y (de multe ori acest număr mai poate fi afișat și ca și un procent). Pentru a utiliza o histogramă în Seaborn putem să utilizăm fie 'sns.displot()' fie 'sns.histplot()'

```python
plt.figure(dpi=200)
sns.displot(x='salary', data=df)
```

În acest moment putem să ne dăm seama în mare parte cam cum sunt împărțite datele respective din moment ce avem și numărul de valori pentru un anumit interval. În ceea ce privește intervalele, acestea se pot modifica utilizând parametrul 'bins'. Acestui pararmetru îi oferim ca și valoare un număr integer care reprezintă numărul de bin-uri pe care dorim să îl avem

```python
plt.figure(dpi=200)
sns.displot(x='salary', data=df, bins=10)
```

```python
plt.figure(dpi=300)
sns.displot(x='salary', data=df, bins=100)
```

Numărul total de bins pe care îl selectăm trebuie să depindă de numărul total de date din coloana respectivă, dar și de ce rezultat dorim să avem, deoarece în funcție de acest parametru graficul arată diferit. De obicei, acest parametru este cel mai des modificat în cadru unei histograme. Ce anume se mai poate modifica pentru un astfel de grafic este stilul (de multe ori, acel fundal alb nu ne ajută să ne dăm seama uneori de numărul de valori dintr-o categorie). Pentru a modifica stilul respectiv se va utiliza codul 'sns.set_style()'. Acesta trebuie apelat înainte de a crea plot-ul iar ca argument o să se ofere un string ce reprezintă numele stilului pe care dorim să îl avem

```python
sns.set_style('darkgrid')
sns.displot(x='salary', data=df)
```

Ce este util la acest plot este faptul că putem seta să ne afișeze KDE-ul. Ce este KDE-ul? KDE-ul este reprezentat de o linie care afișează probabilitatea continuă a unei funcții. Acest KDE provine de la Kernel Density Estimation și este o metodă de estimare a unei funcții de densitate de probabilitate a unei variabile aleatorii. În termeni mai simpli este un mod de estimare printr-o curbă a probabilității continuă a unei variabile finite.

Pentru a afișa acest KDE o să ne folosim de parametrul 'kde' la care o să îi atribuim valoarea True

```python
plt.figure(dpi=300)
sns.displot(x='salary', data=df, kde=True)
```

Să discutăm mai mult despre acest KDE. Pot exista anumite momente când poate dorim să avem afișat doar acest KDE. Pentru astfel de cazuri există metoda 'sns.kdeplot()' pentru care de asemenea trebuie să specificăm DataFrame-ul și pentru ce coloană dorim să facem acest KDE plot

```python
sns.kdeplot(x='salary', data=df)
```
