# Pandas DataFrame - GroupBy Operations Part Two

În partea a doua a tutorialului despre metoda groupby() o să ne focusăm în continuare pe partea de multiindex, pe partea de selectare de date din cadrul unui DataFrame care are un multiindex 

```python
# importing the libraries
import numpy as np
import pandas as pd
```

```python
# reading the csv file into a DataFrame
df = pd.read_csv('../data/03-Pandas/mpg.csv')
```

```python
# printing the first five rows of the DataFrame
df.head()
```

```python
# printing the shape of the DataFrmae
df.shape
```

```python
# creating a DataFrame{ based on groupby() method
year_cly = df.groupby(['model_year', 'cylinders']).mean()
```

```python
# printing the resulting DataFrame
year_cly
```

În continuare o să ne uităm peste procedeul de 'cross section' care reprezintă un procedeu mai general de a extrage anumite date din cadrul unui DataFrame cu un multiindex. Metoda respectivă poartă denumirea de '.xs()'

Metoda '.xs()' are două argumente principale, și anume argumentul de 'key' la care trebuie să îi oferim ca și valoare label-ul pe care dorim să îl extragem și argumentul de 'level' care face referire la nivelul de index. Nivelul de index se poate afla utilizând comanda 'year_cly.index.levels'. Dacă dorim să extragem valorile pentru index-ul 70, atunci la key îi oferim valoarea 70, iar la level îi atribuim valoarea 'model_year'

```python
year_cly.xs(key=70, level='model_year')
```

De notat faptul că această comandă nu acceptă mai mult de o valoare pentru argumentul de key. Dacă dorim să extragem mai multe valori pentru un label de index atunci trebuie să utilizăm indexatorul 'loc'.

```python
year_cly.xs(key=(70, 80), level='model_year')
```

Dacă îi oferim mai multe valori pentru key, atunci o să ridice o eroare. Pentru a extrage datele din label-ul 70 și 80 (ce aparțin de index-ul 'model_year') atunci o să utilizăm indexatorul 'loc'

```python
year_cly.loc[[70, 80]]
```

Din moment ce nu putem extrage mai multe valori probabil vine întrebarea la ce este de folos această metodă? Ceea ce nu am reușit până în acest moment este să extragem valori din DataFrame utilizând index-ul interior (adică 'cylinders'). Poate dorim să extrage doar valorile pentru mașinile cu 4 cilindri, dar dorim să facem asta pentru toți anii din acest DataFrame. Utilizând indexatorul 'loc' nu putem să extragem aceste date, de aceea trebuie să utilizăm această metodă de cross section

```python
year_cly.xs(key=4, level='cylinders')
```

În cadrul codului de mai sus am extras valoarea '4' din cadrul unui index, dar am specificat ca nivelul index-ului să fie 'cylinders', care este un index interior, acesta nu se putea extrage separat utilizând doar indexatorul 'loc', se putea extrage numai împreună cu index-ul exterior și se putea face asta doar pe rând (adică luând fiecare index de 'model_year' în parte)

Să aruncăm o privire din nou peste DataFrame-ul cu datele grupate

```python
year_cly
```

Putem observa că în acest DataFrame avem date la cilindri cu valorile 3, 4, 5, 6, 8. Să zicem că dorim să extragem doar valorile pentru mașinile care au 6 sau 8 cilindri pentru datele grupate. Pentru a selecta aceste date este un procedeu mai complicat, de aceea se indică să se filtreze datele înainte să se grupeze. De exemplu este mult mai simplu să filtrăm datele înainte să le grupăm decât să încercăm să facem acuma această selecție. Din acest motiv este recomandat să se știe de ce date avem nevoie în final pentru a putea face filtrarea respectivă la început astfel încât să avem de-a face doar cu datele de interes.

```python
df[df['cylinders'].isin([6, 8])].groupby(['model_year', 'cylinders']).mean()
```

În continuare putem să arucăm o privire peste modul în care schimbăm nivelele de index și cum putem să sortăm acest multiindex. În cazul de față, index-ul exterior este cel de 'model_year' și cel interior de 'cylinders'.

```python
year_cly
```

În cazul în care nu s-a făcut gruparea în ordinea în care dorim se poate schimba nivelul acestor indexi utilizând nmetoda 'swaplevel()'

```python
year_cly.swaplevel()
```

În cazul de față are mai mult sens să avem ca și prim index model_year deoarece datele din DataFrame arată mai aranjate în acest fel, după cum se vede pentru situația în care avem ca și prim index cilindrii, nu se prea înțelege ce anume reprezintă acești indexi

Pentru a sorta unul dintre indexi prezenți o să se utilizeze tot comanda 'sort_index()' ca și pentru sortarea unui singur index, însă o să îi mai oferim un argument în plus, cel de 'level' prin care practic se specifică ce index dorim să fie sortat.

```python
year_cly.sort_index(level='model_year',ascending=False)
```

În cazul în care se sortează în acest mod utilizând aceași metodă dar pentru index-ul interior, atunci o să se strice acea relație dintre an și index, o să sorteze datele doar după index-ul respectiv, nu ține cont și de index-ul interior. Dacă sortăm după index-ul de 'cylinders', atunci o să treacă prima dată toți indexii care au cea mai mare valoare din cadrul tuturor anilor

```python
year_cly.sort_index(level='cylinders', ascending=False)
```

Până în acest moment am văzut cum anume putem să apelăm o metodă de agregare pentru un DataFrame. Ce anume se poate face atunci când dorim să apelăm mai multe funcții de agregare pentru un DataFrame. De exemplu să zicem că dorim ca pentru DataFrame-ul mare (df) dorim să apelăm metoda 'mean()' și metoda 'std()'. Există o modalitate de a aplea aceste două funcții doar printr-o linie de cod. Acest lucru se face utilizând metoda 'agg()'. Acestei metode o să îi oferim ca și argument o listă care cuprinde numele funcțiilor de agregare pe care dorim să le apelăm. De notat faptul că aceste nume de funcții o să fie trecute ca și string și fără setul de paranteze drepte. Să apelăm cele două funcții pentru DataFrame-ul mare 

```python
df.agg(['mean', 'std'])
```

Ceea ce rezultă este un DataFrame unde ca și index avem numele metodei care s-a apelat, iar ca și valori avem rezultatul acestor metode. Fiind returnat un DataFrame putem să extragem informațiile doar din anumite coloane

```python
df.agg(['mean', 'std'])[['mpg', 'weight']]
```

În DataFrame-ul mare există anumite coloane pentru care nu ar avea sens să apelăm metoda 'std()' deoarece nu are nicio logică rezultatul obținut. Pentru aceste cazuri, cu ajutorul metodei 'agg()' putem să specificăm pentru coloane ce metode să se ruleze. Pentru asta trebuie să oferim ca și argument un dicționar. În cadrul acestui dicționar partea de key reprezintă numele coloanei, iar partea de value reprezintă numele funcțiilor pe care dorim să le apelăm pentru acea coloană

```python
df.agg({'mpg': ['max', 'mean'], 'weight': ['mean', 'std']})
```

Metoda de mai sus reprezintă o metodă mai avansată de a apela anumite metode de agregare pentru anumite coloane. De notat însă faptul că pentru coloanele sau rândurile care nu au la comun aceeași metodă, atunci în acele cazuri o să fie trecută valorea NaN.
