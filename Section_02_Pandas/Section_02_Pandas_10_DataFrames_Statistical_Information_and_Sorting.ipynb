# Pandas DataFrame - Statistical Information and Sorting

În cadrul acestei părți din seria de metode utile din Pandas o să ne uităm peste câteva metode generale pentru partea de descriere sau pentru sortarea datelor. Multe dintre aceste metode sunt ușor de înțeles numai după numele pe care îl au. Pentru a putea utiliaza aceste funcții o să avem nevoie să citim din nou acel fișier csv cu care am lucrat până în acest moment

```python
# importing the libraries
import numpy as np
import pandas as pd
```

```python
# reading the csv file into a DataFrame
df = pd.read_csv('../data/03-Pandas/tips.csv')
```

```python
# showing the first five rows of the DataFrame
df.head()
```

Una dintre aceste metode generale este 'describe()', metodă cu care am mai avut de-a face până acuma. MEtoda respectivă ne returnează o descriere generală a metodelor statistice pentru coloanele ce conțin date numerice

```python
df.describe()
```

O întrebare care apare des este cea de cum putem sorta un DataFrame. Pentru asta o să ne folosim de metoda 'sort_values()', metodă la care îi oferim ca și argument numele coloanei (sub formă de string) după care dorim să facem sortarea respectivă. Un alt argument ce putem să îl utilizăm pentru această metodă este argumentul 'ascending'. Acest argument este setat automat la True, lucru care face să se sorteze datele în mod crescător. Dacă dorim să avem datele în mod descrescător, atunci o să îi atribuim valoarea False

```python
df.sort_values('tip', ascending=False)
```

Un DataFrame poate fi sortat după mai multe coloane, iar în acel caz trebuie să oferim numele coloanelor sub formă de listă (precum facem operația de selectare de coloane multiple, la fel se va face și această precizare de coloane pentru sortare)

```python
df.sort_values(['tip', 'size'])
```

Codul de mai sus prima dată sortează datele după coloana 'tip'. În cazul în care sunt valori duplicate în cadrul acelei coloane, atunci următoarea sortare se va face după coloan 'size'. Ambele sortări o să fie în ordine crescătoare, asta din moment ce nu s-a specificat argumentul 'ascending'. În cazul în care sunt mai multe coloane pe care dorim să le sortăm, putem să specificăm pentru fiecare în parte dacă să se sorteze crescător sau descrescător. Pentru a face asta, la argumentul 'ascending' o să îi atribuim o listă de valori booleane, fiecare valoarea corespunzând pentru o coloană. Prima valoare booleeană corespunde pentru prima coloană selectată și tot așa

```python
df.sort_values(['tip', 'size'], ascending=[True, False])
```

Pentru a afla valoarea maximă dintr-o anumită coloană se va folosi metoda 'max()'. În cazul în care dorim să aflăm poziția (index-ul) pe care se găsește cea mai mare valore, atunci pentru asta există o metodă specială denumită 'idxmax()'

```python
df['tip'].max()
```

```python
df['tip'].idxmax()
```

```python
df.loc[170]
```

```python
df.loc[df['tip'].idxmax()]
```

Pentru valoarea minimă, denumirile funcțiilor sunt 'min()' și 'idxmin()'

O metodă utilă este prin care putem să numărăm valorile unice din cadrul unei coloane. Metoda respectivă poartă denumirea de 'value_counts()'. Această metodă este recomandat să se folosească cu coloane care au anumite categorii prestabilite

```python
df.head()
```

Ce vreau să zic prin anumite categorii prestabilite? De exemplu, în DataFrame-ul de mai sus avem coloana 'sex' care are doar două variante de răspuns. Aceasta este o coloană cu categorii prestabilite, este o coloană care are răspunsurile prestabilite de la început, iar o persoană doar trebuie să aleagă unul dintre acele răspunsuri. Pentru această coloană dorim să aflăm câte valori sunt trecute ca și 'Male' și câte sunt trecute ca și 'Female' 

```python
df['sex'].value_counts()
```

Un alt pas important pentru un DataFrame este să vedem ce valori unice sunt în cadrul unei coloane, sau numărul total de valori unice dintr-o anumită coloană. Pentru a afla valorile unice din cadrul unei coloane, atunci o să folosim metoda 'unique()'. O să folosim această metodă pentru coloana 'day' pentru a afla toate zilele prezente în cadrul acestui DataFrame

```python
df['day'].unique()
```

Pentru a afla numărul total de valori unice, atunci o să folosim metoda 'nunique()'

```python
df['day'].nunique()
```

În continuare o să ne uităm peste anumite metode prin care putem schimba sau înlocui anumite date din cadrul unui DataFrame. Există două metode peste care o să ne uităm, acestea fiind 'replace()' și 'map()'. Metodele sunt destul de asemănătoare, însă una dintre ele folosește dicționarii, iar alta folosește o listă de valori.

```python
df.head()
```

Pentru început o să ne uităm peste metoda 'replace()'. Din cadrul DataFrame-ului de mai sus să presupunem că dorim să modificăm datele din coloana 'sex'. Pentru valorile 'Male' dorim să schimbăm doar cu litera 'M', iar pentru valorea 'Female' dorim să înlocuim această valoare cu litera 'F. Să ne uităm cum putem să modificăm doar una dintre aceste valori. Metoda 'replace()' are două argumente, to_replace= și value=. to_replace reprezintă valorea pe care dorim să o modificăm, iar value reprezintă noua valoare pe care dorim să o modificăm. Această metodă trebuie să o utilizăm pentru un Series, adică pentru coloana unde dorim să facem aceste modificări.

```python
df['sex'].replace(to_replace='Female', value='F')
```

Pentru a modifica mai multe valori, o să folosim o listă de valori, atât pentru argumentul to_replace= cât și pentru argumentul value= (numele acestor argumente pot fi omise)

```python
df['sex'].replace(['Female', 'Male'], ['F', 'M'])
```

Acesta este comportamentul metodei 'replace()'. Se recomandă să se utilizeze această metodă atunci când avem de schimbat o sinfură valore din cadrul unui DataFrame. Se poate utiliza pentru a înlocui rapid valori nule cu o anumită valore, un 0, sau cu ce valoare se dorește să se înlocuiască

În continuare o să ne uităm peste metoda 'map()'. Primnul lucru pe care trebuie să îl facem înainte de a utiliza metoda 'map()' este să ne creem o așa zisă mapă, adică un dicționar care mapează anumite valori. Cheia din dicționarii este valoarea pe care dorim să o modificăm, iar valorea o să fie reprezentată de noua valoare pe care dorim să o atribuim acelei valori.

```python
my_map = {'Female': 'F', 'Male': 'M'}
```

Acest dicționar o să fie trecut după în cadrul metodei 'map()'. Metoda repsectivă trebuie apelată pentru coloana (Series) unde dorim să facema cele modificări

```python
df['sex'].map(my_map)
```

Metoda 'map()' se recomandă să se utilizeze atunci când se dorește să se înlocuiască mai multe valori din cadrul unui DataFrmae, această metodă fiind mai ușor de citit.

De multe ori lucrând cu un set foarte mare de date există o posibilitate ca unele rânduri din cadrul unui DataFrame să fie duplicate. Pentru a afla locurile unde există rânduri duplicate putem să utilizăm metoda 'duplicated()'. Metoda respectivă returnează un Series de valori Booleane unde valoare True înseamnă că este un rând duplicat, iar False înseamnă că acel rând nu este duplicat. Dacă există un anumit rând duplicat, atunci o să returneze True pentru prima instanță a rândului duplicat.

```python
df.duplicated()
```

```python
simple_df = pd.DataFrame([1, 2, 2, 2, 3, 2], ['a', 'b', 'c', 'd', 'e', 'f'])
```

```python
simple_df
```

```python
simple_df.duplicated()
```

Mai sus am creat un DataFrame extrem de simplu pentru a vedea mai exact ce anume returnează metoda 'duplicated()'. Din moment ce găsește o valore duplicată returnează True pentru acel rând și pentru toate celelalte rânduri care mai urmează în cazul în care acestea sunt duplicate. Pentru a șterge valorile duplicate o să folosim metoda 'drop_duplicates()'

```python
simple_df.drop_duplicates()
```

O altă metodă ce poate fi extrem de utilă uneori este metoda 'between()'. Această metodă returnează toate valorile din cadrul unei coloane care se găsesc în intervalul a două valori alese de utilizator.

```python
df['total_bill']
```

Mai sus avem datele din coloana 'total_bill'. Din această coloană dorim să extragem datele care au valori cuprinse între 10 și 20. Pentru asta o să apelăm metoda 'between()', căreia îi oferim ca și argument cele două valori, 10 și 20. Extremele acestor valori (adică 10, respectiv 20) nu sunt incluse în cadrul outputului. Dacă dorim ca aceste valori să fie incluse, atunci o să utilizăm argumentul 'inlcusive' la care o să îi atribuim valorea 'both'

```python
df['total_bill'].between(10, 20, inclusive='both')
```

Rezultatul este un Series de valori boolean unde True înseamnă că valorea din acel rând se încadrează în acel interval. După cum știm, putem să utilizăm comanda de mai sus ca și un filtru pentru a ne returna un DataFrame doar cu valorile respective

```python
df[df['total_bill'].between(10, 20, inclusive='both')]
```

Am văzut în cadrul acestui tutorial cum anume putem să extragem cea mai mare valore din cadrul unei coloane. Poate evem un caz în care dorim să extragem cele mai mari 5 valori, sau 10 valori. Pentru cazul acesta o să ne folosim de metoda 'nlargest()'. Ca și argumente pentru aceste metodă o să îi atribuim un număr care repezintă numărul de elemente (cele mai mari) pe care dorim să îl extragem împreună cu numele coloanei de unde dorim să facem această extragere. Metoda de mai sus o să ne extragă tot DataFrame-ul, nu doar valorie din coloana pe care am pasat-o ca și argument

```python
df.nlargest(10, 'tip')
```

Pentru a extrage valorile cele mai mici, se va utiliza metoda 'nsmallest()'

Metoda de mai sus este rezultatul combinării a două metode individuale. Putem prima dată să sortăm aceste date după valorile din coloane 'tip', iar după să apelăm metoda 'head()' pentru a ne extrage un anumit număr de rânduri din acest DataFrame

```python
df.sort_values('tip', ascending=False).head(10)
```
