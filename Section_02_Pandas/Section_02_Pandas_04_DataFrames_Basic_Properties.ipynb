# Pandas DataFrame - Basic Properties

```python
# importing the libraries
import numpy as np
import pandas as pd
```

```python
# reading the csv file
df = pd.read_csv('../data/03-Pandas/tips.csv')
```

```python
# listing the elements inside the data frame
df
```

Până în acest moment am învățat cum anume să citim date dintr-un fișier csv. Având aceste date salvate într-un data frame o să începem să explicăm ce anume se există în acest data frame după care o să trecem prin câteva metode și atribute generale pentru un data frame pentru a putea înțelege mai bine ce anume se găsește în cadrul acestui data frame. 

Acest data frame reprezintă un set de date unde un ospătar a înregistrat informații despre fiecare tip pe care l-a primit pe o perioadă de câteva luni de zile. După cum se observă, în cadrul acestui data frame există un număt de 11 coloane și 244 de rânduri. Aceste coloane conțin informații referitoare la toatul notei de plată, la bacșișul care s-a lăsat, la numărul total de persoane care au fost la masă, sexul persoanei care a plătit și alte informații. Un prim lucru pe care poate dorim să îl aflăm despre acest data frame reprezintă numele coloanelor. Pentru a returna toate denumirile coloanelor putem să ne utilizăm de atributul 'columns'. Acest atribut returnează sub formă de listă toate numele coloanelor

```python
df.columns
```

Numele coloanelor sunt importante deoarece dacă dorim să extragem anumite informații din cadrul unuei singure coloane atunci trebuie să îi oferim numele acelei coloane. De asemena, un atribut destul de important ce îl putem utiliza este 'index'. Acest atribut ne returnează valorile pentru fiecare index din cadrul unui data frame

```python
df.index
```

În cazul de față putem să observăm că acest index reprezintă un range creat automat cu valori de la 0 la 244. Dacă acest data frame avea un index de altă formă (un string label), atunci valorile acestui index ar fi fost returnate sub forma unuei liste precum ceea ce a fost returnat de către atributul columns

Pentru acest data frame avem un număr de 244 de coloane. Dacă dorim să afișăm doar primele rânduri din cadrul unui data frame, atunci putem utiliza metoda 'head()'. Metoda respectivă ne returnează primele 5 rânduri din cadrul unui data frame

```python
df.head()
```

În mod default, metoda 'head()' ne returnează primele 5 rânduri dintr-un data frame. Dacă dorim, în cadrul parantezelor putem să trecem o valoare care o să reprezinte numărul de rânduri care dorim să fie afișat. Dacă dorim să ne afișeze doar primele 2 rânduri din data frame, atunci pentru metoda 'head()' o să îi oferim ca și argument valorea 2

```python
df.head(2)
```

De asemenea există și o metodă prin care putem extrage ultimele valori din cadrul unui data frame, iar acea metodă poartă denumirea de 'tail()'. Metoda respectivă are același comportament precum 'head()', ne retunrează 5 rânduri dacă nu se oferă nicio valoare ca și argument

```python
df.tail()
```

După cum am văzut și anterior, pentru a afla informații generale despre un data frame putem să utilizăm metoda 'info()'. Această metodă ne afișează date generale despre data frame, cum ar fi numărul total de rânduri și de coloane, denumirea fiecărei coloane, tipul de date stocat în cadru fiecărui coloană (de notat faptul că pandas în loc de string trece ca și denumire object. Prin urmare, unde avem object înseamnă că avem de aface cu un tip de date de tipul string)

```python
df.info()
```

După cum se poate observa, avem de a face și cu unele coloane care rețin date de tipul integer sau float. Pentru aceste date putem să executăm anumite funcții care să ne returneze valoare medie din cadrul coloanei, valorea maximă, minimă, deviația standard sau alte valori. Pentru a putea calcula și vizualiza toate aceste date, pandas ne pune la dispoziție o metodă denumită 'descibe()' care face automat acesta calcule și le afișează

```python
df.describe()
```

Dacă dorim să inversăm datele din acel data frame (ca denumirea coloanelor să fie trecută ca și index, respectiv invers) în cazul în care ni se pare că este mai ușor să vizualizăm datele în acel format, atunci putem să utilizăm metoda 'transpose()' care o să realizeze acest lucru

```python
df.describe().transpose()
```
